```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
archivo= pd.read_csv(".\messi_barca.csv", delimiter="," , decimal=",", encoding="latin-1")
archivo
```

```python
#goles por temporada
archivo[["Season"]].astype(str)
archivo[["Season","Goals scored"]].groupby(["Season"]).sum().sort_values(by="Goals scored" , ascending=False)
```

```python
archivo.info()
```

```python
#Temporada con mas minutos jugados
print(archivo.loc[archivo["Minutes played"].idxmax(),["Season"]])
print(archivo["Minutes played"].max())
```

```python
#Goles en Liga con menos de 22 años
archivo[(archivo["Competition"]=="La Liga") & (archivo["Age"]<=21)].sum()[["Goals scored"]]
```

```python
archivo.corr()
```

```python
plt.hist(archivo.Starts, bins= 6)
plt.xlabel('goles')
plt.grid()
plt.show()
```
